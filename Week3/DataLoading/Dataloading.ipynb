```python
import pandas as pd
import sqlite3
from loguru import logger

doel_conn = sqlite3.connect('../../Source Data Model/BikeToDrive_V3_OLTP.db', timeout=30)
doel_cursor = doel_conn.cursor()
```

Accessoire_Inkoop

```python
source_conn = sqlite3.connect('../../Databronnen/BikeToDrive_4_Accessoire_Inkoop.db')
source_cursor = source_conn.cursor()

logger.info("Start dataloading Accessoire_Inkoop")

try:
    # Delete voor Full-Refresh
    alle_tabellen = [
        'Accessoire_Inkoop_Leverancier',
        'Accessoire_Inkoop_Accessoire',
        'Accessoire_Inkoop'
    ][::-1] # We draaien de volgorde van de tabellen om.

    # We deleten alles van onze bestaande OLTP database
    for tabel in alle_tabellen:
        try:
            doel_cursor.execute(f"DELETE FROM {tabel};")
            logger.info(f"Tabel geleegd: {tabel}")
        except Exception as e:
            logger.error(f"FAILED DELETE on {tabel}: {e}")

    doel_conn.commit()
    logger.info("Delete afgerond") # loggen succes delete

    # Fetchen data van leveranciers en loggen hiervan
    source_cursor.execute("SELECT * FROM Leverancier")
    leveranciers = source_cursor.fetchall()
    logger.info(f"{len(leveranciers)} leveranciers opgehaald")

    for row in leveranciers:
        try:
            doel_cursor.execute(
                "INSERT INTO Accessoire_Inkoop_Leverancier VALUES (?, ?, ?, ?)", # id, naam, adres, woonplaats
                row
            )
        except Exception as e:
            logger.error(f"Insert failed Leverancier: {row} ({e})")

    doel_conn.commit()
    logger.info(f"{len(leveranciers)} leveranciers geladen")

    # Fetchen data van accessoires en loggen hiervan
    source_cursor.execute("SELECT * FROM Accessoire")
    accessoires = source_cursor.fetchall()
    logger.info(f"{len(accessoires)} accessoires opgehaald")

    for row in accessoires:
        try:
            doel_cursor.execute(
                "INSERT INTO Accessoire_Inkoop_Accessoire VALUES (?, ?, ?, ?, ?, ?)", # id, naam, standaardprijs, inkoopprijs, soort, leverancier
                row
            )
        except Exception as e:
            logger.error(f"Insert failed Accessoire: {row} ({e})")

    doel_conn.commit()
    logger.info(f"{len(accessoires)} accessoires geladen")

    # Fetchen data van inkoop en loggen hiervan
    source_cursor.execute("SELECT * FROM Accessoire_Inkoop")
    inkoop = source_cursor.fetchall()
    logger.info(f"{len(inkoop)} inkoop opgehaald")

    for row in inkoop:
        try:
            doel_cursor.execute(
                "INSERT INTO Accessoire_Inkoop VALUES (?, ?, ?, ?, ?)", # id, maand, jaar, aantal, accessoire
                row
            )
        except Exception as e:
            logger.error(f"Insert failed Accessoire_Inkoop: {row} ({e})")

    doel_conn.commit()
    logger.info(f"{len(inkoop)} inkoopregels geladen")

    logger.info("dataloading succesvol afgerond")

except Exception as e:
    logger.critical(f"dataloading volledig gefaald: {e}")

finally:
    # sluiten connecties
    source_cursor.close()
    doel_cursor.close()
    source_conn.close()
    doel_conn.close()

    logger.info("Database connecties gesloten")
```

```python
# We maken eerst connectie met de bron.
source_conn = sqlite3.connect('../../Databronnen/BikeToDrive_2_Fietsverkoop.db')
source_cursor = source_conn.cursor()

# Lijst met alle tabellen die geleegd moeten worden.

try:
    
    logger.info("Start Data Loading Fietsverkoop")
    alle_tabellen = [
        'Fiets_Verkoop_Fabrikant',
        'Fiets_Verkoop_Filiaal',
        'Fiets_Verkoop_Fiets',
        'Fiets_Verkoop_Monteur',
        'Fiets_Verkoop_Klant',
        'Fiets_Verkoop'
    ][::-1] # We draaien de volgorde om.

# We deleten alles van onze bestaande OLTP database
    for tabel in alle_tabellen:
        try:
            doel_cursor.execute(f"DELETE FROM {tabel}")
            logger.info(f"Tabel geleegd: {tabel}")
        except Exception as e:
            logger.error(f"FAILED DELETE on {tabel}: {e}")

# Commit na delete
doel_conn.commit()
logger.info("Delete afgerond") # loggen succes delete

# De SQL Query die alles selecteert vanuit de Operationele Database.
# We gaan eerst het tabel Fabrikant inserten in onze OLTP.
source_cursor.execute("SELECT * FROM Fabrikant")
Fabrikant = source_cursor.fetchall()
logger.info(f"{len(Fabrikant)} fabrikanten opgehaald")

# 1. Fiets_Verkoop_Fabrikant
for row in Fabrikant:
        try:
            doel_cursor.execute("INSERT INTO Fiets_Verkoop_Fabrikant VALUES (?, ?, ?, ?)", row) # id,  naam,  adres, plaats
        except Exception as e:
            logger.error(f"Insert failed Fiets_Verkoop_Fabrikant: {row} ({e})")
        
    
doel_conn.commit()
logger.info(f"{len(Fabrikant)} fabrikanten geladen")

# 2. Fiets
source_cursor.execute("SELECT * FROM Fiets")
Fiets = source_cursor.fetchall()
logger.info(f"{len(Fiets)} fietsen opgehaald")

for row in Fiets:
    try:
        doel_cursor.execute("INSERT INTO Fiets_Verkoop_Fiets VALUES (?, ?, ?, ?, ?, ?, ?, ?)", row) # id, soort, merk, type, standaardprijs, inkoopprijs, kleur, fabrikant
    except Exception as e:
        logger.error(f"Insert failed Fiets_Verkoop_Fiets: {row} ({e})")

doel_conn.commit()
logger.info(f"{len(Fiets)} fietsen geladen")

# 3. Filiaal
source_cursor.execute("SELECT * FROM Filiaal")
Filiaal = source_cursor.fetchall()
logger.info(f"{len(Filiaal)} filialen opgehaald")

for row in Filiaal:
    try:
        doel_cursor.execute("INSERT INTO Fiets_Verkoop_Filiaal VALUES (?, ?, ?, ?)", row) # id, naam, adres, provincie
    except Exception as e:
        logger.error(f"Insert failed Fiets_Verkoop_Filiaal: {row} ({e})")

doel_conn.commit()
logger.info(f"{len(Filiaal)} filialen geladen")

# 4. Monteur
source_cursor.execute("SELECT * FROM Monteur")
Monteur = source_cursor.fetchall()
logger.info(f"{len(Monteur)} monteurs opgehaald")

for row in Monteur:
    try:
        doel_cursor.execute("INSERT INTO Fiets_Verkoop_Monteur VALUES (?, ?, ?, ?, ?)", row) # id, naam, woonplaats, uurloon, filiaal
    except Exception as e:
        logger.error(f"Insert failed Fiets_Verkoop_Monteur: {row} ({e})")

doel_conn.commit()
logger.info(f"{len(Monteur)} monteurs geladen")

# 5. Klant
source_cursor.execute("SELECT * FROM Klant")
Klant = source_cursor.fetchall()
logger.info(f"{len(Klant)} klanten opgehaald")

for row in Klant:
    try:
        doel_cursor.execute("INSERT INTO Fiets_Verkoop_Klant VALUES (?, ?, ?, ?, ?, ?)", row) # id, naam, adres, woonplaats, geslacht, geboortedatum
    except Exception as e:
        logger.error(f"Insert failed Fiets_Verkoop_Klant: {row} ({e})")

doel_conn.commit()
logger.info(f"{len(Klant)} klanten geladen")

# 6. Verkoop
source_cursor.execute("SELECT * FROM Fiets_Verkoop")
Verkoop = source_cursor.fetchall()
logger.info(f"{len(Verkoop)} verkoop opgehaald")

    for row in Verkoop:
        try:
        doel_cursor.execute("INSERT INTO Fiets_Verkoop VALUES (?, ?, ?, ?, ?, ?, ?)", row) # id, datum, aantal, verkoopprijs, klant, fiets, monteur
        except Exception as e:
        logger.error(f"Insert failed Fiets_Verkoop: {row} ({e})")

    doel_conn.commit()
    logger.info(f"{len(Verkoop)} verkoop geladen")

except Exception as e:
    logger.critical(f"dataloading volledig gefaald: {e}")


# We committen onze gewijzigde acties.
finally:
doel_conn.commit()
source_cursor.close()
source_conn.close()
doel_cursor.close()
doel_conn.close()

logger.info("Database connecties gesloten")
```

Toevoegen van een extra rij

```python
# We voegen een extra rij toe aan de Operationele Database (Fietsverkoop)

# We maken eerst de connectie met de Operationele Database
source_conn = sqlite3.connect('../../Databronnen/BikeToDrive_2_Fietsverkoop.db')
source_cursor = source_conn.cursor()

# We voegen een extra rij toe aan het tabel van Fiets.

sql_insert = """
INSERT INTO Fiets 
VALUES (1500, 'Stadsfiets', 'Gazelle', 'Classic', 599.99, 400.00, 'Zwart', 1)
"""

source_cursor.execute(sql_insert)
source_conn.commit()

source_cursor.close()
source_conn.close()

```

Wijzigen van de extra rij

```python
# We wijzigen deze rij in de Operationele Database.

# We maken eerst de connectie met de Operationele Database
source_conn = source_conn = sqlite3.connect('../../Databronnen/BikeToDrive_2_Fietsverkoop.db')
source_cursor = source_conn.cursor()

# We wijzigen de rij van kleur zwart naar kleur blauw
sql_update = """
UPDATE Fiets
SET kleur = 'Blauw'
WHERE fietsnr = 1500
"""

source_cursor.execute(sql_update)
source_conn.commit()

source_cursor.close()
source_conn.close()
```

```python
# We verwijderen de rij in de Operationele Database.

# We maken eerst de connectie met de Operationele Database
source_conn = source_conn = sqlite3.connect('../../Databronnen/BikeToDrive_2_Fietsverkoop.db')
source_cursor = source_conn.cursor()

# We deleten de fiets met PK = 1500
sql_delete = """
DELETE FROM Fiets
WHERE fietsnr = 1500
"""

source_cursor.execute(sql_delete)
source_conn.commit()

source_cursor.close()
source_conn.close()
```
